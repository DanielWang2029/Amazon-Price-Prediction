```python
import json
import pandas as pd
import numpy
```

```python
# 1. load and convert
data = []
with open('meta_Appliances.json') as f:
    for l in f:
        data.append(json.loads(l.strip()))
        
# 2. Convert into df
df = pd.DataFrame.from_dict(data)
```

```python
# 3. Data Cleaning

# Clean non valid entries
df = df.fillna('')
df = df[~df.title.str.contains('getTime')]
len(df)
```

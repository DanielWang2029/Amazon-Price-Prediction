```python
import json
import pandas as pd
import numpy
```

```python
# 1. load and convert
data = []
with open('data/meta_Appliances.json') as f:
    for l in f:
        data.append(json.loads(l.strip()))
        
# 2. Convert into df
df = pd.DataFrame.from_dict(data)
```

```python
# 3. Data Cleaning

# Clean non valid entries
df = df.fillna('')
df = df[~df.title.str.contains('getTime')]
len(df)
```

```python
df
```

```python
df1 = df[~df.price.str.contains('#')]
df2 = df1[df1['price'] != '']
# df3 = df2[df2['also_view'].map(len) != 0]
# df3 = df3[df3['also_buy'].map(len) != 0]
df3 = df2
```

```python
df4 = df3.drop(['imageURL', 'imageURLHighRes', 'date','details','fit','tech2','tech1','similar_item'], axis = 1)
```

```python
df4
```

```python
from re import sub
from decimal import Decimal

def money2float(money):
    try:
        return float(Decimal(sub(r'[^\d.]', '', money)))
    except:
        return None

df4['price'] = df4['price'].map(money2float)
```

```python
cleanedDF = df4
cleanedDF
```

```python
cleanedDF = cleanedDF.drop_duplicates(subset = 'asin')
```

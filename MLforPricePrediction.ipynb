```python
import json
import pandas as pd
import numpy
import torch
import sklearn
```

```python
mother = pd.read_json('cleanedAppliances.json')
```

```python
# With Mother datapool, we can extract the subset
target = mother[(mother['also_buy'].map(len) != 0) & (mother['also_view'].map(len) != 0)]
```

```python
# list[asin] -> list[price]
def asin2price(asins):
    prices = []
    for asin in asins:
        price = mother.loc[mother["asin"] == asin].price.values
        if len(price) == 0:
            prices.append(0)
        else:
            prices.append(price[0])
    return prices

asin2price(["B00Q9EJODK", "B000HMBWUI", "B00024JWG2", "B00TCXN1LQ"])
```
